```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000

x = np.linspace(-10,10, num=N)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

y = g(x)

f_o_g = f(y)

plt.plot(x,f_o_g)
plt.show()
```

```python
N = 1000

x = np.linspace(-10,10, num=N)

def h(x):
    return x**2

def g(x):
    return np.tan(x)

def f(x):
    return np.log2(g(x))

y = g(h(x))

f_o_g_o_h = f(g(h(x)))

plt.figure(figsize=(10,7))
plt.plot(x,f_o_g_o_h, 'r-')
plt.show()
```

```python

N = 1000

x = np.linspace(-10,10, num=N)

def g(x):
    return np.sin(x)

def f(x):
    return np.tan(x)

def absoluto(x):
    y = np.absolute(x)
    return y

def logaritmico(x):
    return np.log2(x)


y = logaritmico(x)

f_o_g = absoluto(y)

plt.plot(x,f_o_g)
plt.show()
```
